```python
import pandas as pd
dataset=pd.read_csv("TradeSettlementDetails.csv",index_col=None)
dataset
```

```python
dataset.info()
```

```python
independent=dataset[['Trade Value', 'Trade Volume', 'Counterparty ID', 'Counterparty Risk Score',
 'Counterparty Failures', 'Market Volatility', 'Liquidity',
 'Manual Intervention', 'Client ID', 'Settlement Type_FoP']]
dependent=dataset["Settlement Status_Success"]
```

```python
#split into training set and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(independent, dependent, test_size = 1/3, random_state = 0)
X_train
X_test
y_train
y_test
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
X_train
X_test
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

param_grid={'kernel':['poly','rbf','sigmoid'],
           'gamma':['scale','auto'],
           'max_iter':[1, 5],
           'class_weight': ['None','balanced']}

grid=GridSearchCV(SVC(probability=True),param_grid,refit=True,verbose=3,n_jobs=-1,scoring='f1_weighted')
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
re=grid.cv_results_
print(re)
```

```python
table=pd.DataFrame.from_dict(re)
table
```

```python
Y_predict=grid.predict(X_test)
Y_predict
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import roc_auc_score
cm = confusion_matrix(y_test, Y_predict)
clf_report = classification_report(y_test, Y_predict)
ras=roc_auc_score(y_test,grid.predict_proba(X_test)[:,1])
print(ras)
print(cm)
print(clf_report)
```
