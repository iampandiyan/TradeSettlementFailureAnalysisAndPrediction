```python
reason_abbreviation = {
    'Insufficient Funds': 'INSFND',
    'Invalid Counterparty': 'INVCTR',
    'Technical Error': 'TECHER',
    'Market Closure': 'MRKCLS',
    'Documentation Issue': 'DOCISS',
    'Insufficient Funds at the Counterparty': 'INSCPT',
    'Issues at the CSD': 'CSDISS',
    'Client Instruction Missing': 'CLIMIS',
    'Client Short to Deliver': 'CLISHD',
    'Counterparty Short to Deliver': 'CPTSHD',
    'Currency Difference': 'CURDIF',
    'Account Difference': 'ACCDIF',
    'Quantity Difference': 'QTYDIF'
}

dataset['Settlement Failure Reason'] = dataset['Settlement Failure Reason'].replace(reason_abbreviation)
```
