# Test the Analysis of Variance between Processing Time and Trade Volum at signifance level 5%.(Make decision using Hypothesis Testing)


```python
#One Way Classification
import scipy.stats as stats
f_statistic,p_value=stats.f_oneway(dataset['Counterparty Failures'],dataset['Counterparty Failure Rate'])
print("F-statistic:", f_statistic)
print("p-value:", p_value)
H0='There is no significance difference between the Processing Time and Trade Volume'
H1='There is a significance difference between  the Processing Time and Trade Volume'
if p_value < 0.05:
    print(H1)
else:
    print(H0)
```

# Test the similarity between the Settlement Status(Failure) and Settlement Failure Reason(Client Instruction MissingR) with respect to Processing Time at significance level of 5%.(Make decision using Hypothesis Testing)


```python
from scipy.stats import ttest_ind
ss=dataset[dataset['Settlement Status']=='Fail']['Processing Time']
sfr=dataset[dataset['Settlement Failure Reason']=='Client Instruction Missing']['Processing Time']
stat,pval=ttest_ind(ss,sfr)
print(stat)
print(pval)
H0='There is no similarity between the Settlement Status(Fail) and Settlement Failure Reason(Client Instruction Missing) with respect to Processing Time at significance level of 5%'
H1='There is similarity between the Settlement Status(Fail) and Settlement Failure Reason(Client Instruction Missing) with respect to Processing Time at significance level of 5%'
if(pval<0.05):
    print(H1)
else:
    print(H0)
```
